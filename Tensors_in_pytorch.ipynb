A tensor is simply a generalized form of data used in deep learning to store and process information.

Why are Tensors Needed in Deep Learning?
Deep learning works with complex data:
 - Images
 - Text
 - Audio
 - Video
 - Sensor data

All of these naturally require multiple dimensions, and tensors are the most efficient way to handle them.

### **Creating a Tensor**

```python
# Using empty
torch.empty(2,3)
```

```python
# Using zeros/ones
torch.ones(4,6)
```

```python
torch.zeros(2,3)
```

```python
# Using random entries
torch.rand(5,9)
```

```python
# Custom Tensor
torch.tensor([12,2,3])
```

### **Mathematical Operations on tensors**

```python
x = torch.rand(2,3)
y = torch.rand(2,3)
z = torch.rand(3,2)
k = torch.rand(4,4)
```

```python
x
```

```python
x + 2
```

```python
x**2
```

```python
# Matrix Multiplication
torch.matmul(x,z)
```

```python
# Determinant of a Matrix
torch.det(k)
```

```python
# Comparisons

print(x > y)
print(x >= y)
print(x <= y)
print(x == y)
print(x != y)
```

```python
torch.sum(x)
```

```python
torch.max(y)
```

```python
torch.argmax(y)
```

```python
torch.log(x)
```

```python
torch.sigmoid(x)
```

```python
torch.relu(k)
```

```python
x = x + y
```
