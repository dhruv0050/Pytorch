```python
import torch
```

Let's find dy/dx for a given value of x

```python
x = torch.tensor(3.0, requires_grad = True)
```

```python
y = x**2
```

```python
y.backward()
```

```python
x.grad
```

Let's find dz/du where z is a function of  v and v is a function of u

```python
u = torch.tensor(2.0, requires_grad = True)
```

```python
v = u**3 + u**2
```

```python
z = torch.sin(v)
```

```python
z.backward()
```

```python
u.grad
```

Clearing gradients : In PyTorch, clearing gradients means resetting stored gradients (usually to zero) before the next backward pass.

```python
y = x**2
y.backward()
print(x.grad)
```

```python
y = x**2
y.backward()
print(x.grad)
```

So this may cause issue while training. Hence it needs to be fixed

```python
x.grad.zero_()
```
